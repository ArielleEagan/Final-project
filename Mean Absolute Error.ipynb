```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
fpath = 'salaries_modified.csv'
fpath2 = 'salary.csv'
```

```python
df_cleaned = pd.read_csv(fpath)
df = pd.read_csv(fpath2)
df
```

```python
mean = df_cleaned.totalyearlycompensation.mean()
```

```python
mean
```

```python
df['mean'] = mean
df.head()
```

```python
y_true = df.totalyearlycompensation
y_pred = df["mean"]
```

```python
MAE = mean_absolute_error(y_true,y_pred)
MAE
```
